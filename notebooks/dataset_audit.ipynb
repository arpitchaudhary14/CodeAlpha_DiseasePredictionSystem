# MediSense Dataset Audit Report

This notebook automatically discovers and analyzes all CSV files present in the `datasets` folder.

```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown

# Settings
DATASETS_DIR = '../datasets'
sns.set_theme(style="whitegrid", palette="muted")
pd.set_option('display.max_columns', None)
```

## 1. Dataset Discovery

```python
# Automatically discover CSV files recursively
csv_files = glob.glob(os.path.join(DATASETS_DIR, '**/*.csv'), recursive=True)
display(Markdown(f"**Found {len(csv_files)} datasets:**"))
for f in csv_files:
    print(f" - {f}")
```

## 2. Dataset Analysis and Basic EDA
For each dataset, we will load the data and perform a basic analysis.

```python
dataset_stats = []

for file_path in csv_files:
    dataset_name = os.path.basename(file_path)
    display(Markdown(f"---\n### Dataset: `{dataset_name}`"))
    
    # Load dataset
    try:
        df = pd.read_csv(file_path)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        continue
        
    # Shape
    display(Markdown(f"**Shape:** {df.shape[0]} rows, {df.shape[1]} columns"))
    
    # Columns & Dtypes
    display(Markdown("**Columns & Data Types:**"))
    dtype_df = pd.DataFrame(df.dtypes, columns=['Data Type']).reset_index().rename(columns={'index': 'Column'})
    display(dtype_df.head(10))
    if len(dtype_df) > 10:
        print(f"... and {len(dtype_df) - 10} more columns")
        
    # Missing values analysis
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    if missing.empty:
        display(Markdown("**Missing Values:** None"))
    else:
        display(Markdown("**Missing Values:**"))
        display(pd.DataFrame(missing, columns=['Missing Count']))
        
    # Duplicate analysis
    duplicates = df.duplicated().sum()
    display(Markdown(f"**Duplicates:** {duplicates} ({duplicates/len(df):.2%} of data)"))
    
    # Class distribution / Categorical Columns
    display(Markdown("**Class Distribution (Categorical/Target Columns):**"))
    numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
    
    potential_targets = [col for col in df.columns if df[col].nunique() <= 10]
    
    if not potential_targets:
        print("No columns with <= 10 unique values found for class distribution.")
    else:
        for col in potential_targets[:5]: # limit to first 5 to avoid clutter
            display(Markdown(f"*{col}*:"))
            display(df[col].value_counts().to_frame())
            
    # Basic EDA - Visualizations
    display(Markdown("**Basic EDA:**"))
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Plot missing values if any, else plot correlation matrix
    if not missing.empty:
        sns.barplot(x=missing.index, y=missing.values, ax=axes[0])
        axes[0].set_title('Missing Values by Column')
        axes[0].tick_params(axis='x', rotation=45)
    else:
        # Correlation matrix for first 10 numeric columns
        corr_cols = numeric_cols[:10]
        if corr_cols:
            sns.heatmap(df[corr_cols].corr(), annot=False, cmap='coolwarm', ax=axes[0])
            axes[0].set_title('Correlation Matrix (Sample)')
            
    # Histogram of a numeric column
    if numeric_cols:
        sns.histplot(df[numeric_cols[0]], bins=30, kde=True, ax=axes[1])
        axes[1].set_title(f'Distribution of {numeric_cols[0]}')
        
    plt.tight_layout()
    plt.show()
    
    # Save stats for final report
    dataset_stats.append({
        'Dataset': dataset_name,
        'Rows': df.shape[0],
        'Columns': df.shape[1],
        'Missing Values': missing.sum(),
        'Duplicates': duplicates
    })
```

## 3. Final Comparison Report

```python
display(Markdown("### Overview"))
if dataset_stats:
    comparison_df = pd.DataFrame(dataset_stats)
    display(comparison_df)
    
    fig, ax = plt.subplots(1, 2, figsize=(14, 5))
    sns.barplot(data=comparison_df, x='Dataset', y='Rows', ax=ax[0])
    ax[0].set_title('Number of Rows per Dataset')
    ax[0].tick_params(axis='x', rotation=45)
    
    sns.barplot(data=comparison_df, x='Dataset', y='Columns', ax=ax[1])
    ax[1].set_title('Number of Columns per Dataset')
    ax[1].tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()
else:
    print("No datasets processed.")
```
